```python
#Are students more likely to respond to a hard course or an easy course?

#import statements and load data ---

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("fce_data.csv")

#drop buggly column 
df.head()
```

```python
df.isna().sum()
```

```python
df = df.drop(columns = ["Semester"])

df_cy = df.groupby(["College", "Year"], as_index=False).mean()
piv = df_cy.pivot("College", "Year", "Hrs Per Week")
ax1 = sns.heatmap(piv, annot=True)
ax1.set_ylim(0, 10)
```

```python
df_cy = df.groupby(["College", "Year"], as_index=False).mean()
piv = df_cy.pivot("College", "Year", "Hrs Per Week 5")
ax1 = sns.heatmap(piv, annot=True)
ax1.set_ylim(0, 10)
```

```python
df_cy = df.groupby(["College", "Year"], as_index=False).mean()
piv = df_cy.pivot("College", "Year", "Hrs Per Week 8")
ax1 = sns.heatmap(piv, annot=True)
ax1.set_ylim(0, 10)
```

```python
old_data = df[df["Year"] < 2014]
old_data.head(10)
```

```python
n_df = df[df["Year"] >= 2014]
n_df = n_df.drop(columns = ["Hrs Per Week 5", "Hrs Per Week 8"])
```

```python
n_df.isna().sum()
```

```python
n_df = n_df.fillna(n_df.mean())
```

```python
ax1 = sns.heatmap(n_df.corr(), annot=True, fmt = '.1f')
ax1.set_ylim(0, 14)
```

```python
#Clean up the data by dropping columns with crazy amounts of nulls and filling in the few remaining gaps w/ the mean
n_df = n_df.drop(columns = ["Semester", "Hrs Per Week 5", "Hrs Per Week 8"])
```
